```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
X,Y=np.meshgrid(np.linspace(-1.57,1.57,100),np.linspace(-1.57,1.57,100))
Z=np.sin(2*X)+1.5*np.cos(Y)**4+2*np.cos(X+Y)
plt.contour(X,Y,Z,50)
X=X[::5,::5]
Y=Y[::5,::5]
VX=2*np.cos(2*X)-2*np.sin(X+Y)
VY=-6*np.cos(Y)**3*np.sin(Y)-2*np.sin(X+Y)
plt.quiver(X,Y,VX,VY)
plt.show()
```

```python
def draw(plt, x, y, beta, gamma):
    xs = [x]
    ys = [y]
    for i in range(1000):
        nxt_x = xs[-1] - beta * xs[-1] * ys[-1]
        nxt_y = ys[-1] + beta * xs[-1] * ys[-1] - gamma * ys[-1]
        xs.append(nxt_x)
        ys.append(nxt_y)
    plt.plot(xs, ys)
```

```python
X, Y = np.meshgrid(np.linspace(0, 1, 100), np.linspace(0, 1, 100))
print(X.shape)
X=X[::5,::5]
Y=Y[::5,::5]

beta = 0.0857
gamma = 0.03
VX = -beta * np.multiply(X, Y)
VY = -VX - gamma * Y

plt.quiver(X,Y,VX,VY)

x = np.linspace(gamma / beta, gamma / beta, 100)
y = np.linspace(0, 1 - gamma / beta, 100)
plt.plot(x, y);

x2 = np.linspace(0, 1, 100)
y2 = np.linspace(1, 0, 100)
plt.plot(x2, y2);

draw(plt, 0.1, 0.9, beta, gamma)
draw(plt, 0.9, 0.1, beta, gamma)
draw(plt, 0.3, 0.7, beta, gamma)
draw(plt, 0.6, 0.4, beta, gamma)

plt.show()
```
